# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

The side-by-side plots make it easier to focus on each week individually. Each graph is less cluttered, which improves readability and allows the viewer to clearly observe trends within a single week. However, this approach makes it harder to directly compare the two weeks because the viewer must mentally align the patterns across the separate plots.

The combined plot, where both weeks are displayed on the same graph, allows for direct comparison between the two time series. It is easier to identify similarities and differences, such as which week had higher visitor counts on specific days. However, the combined plot can become slightly more cluttered, especially if there are many lines, which may reduce clarity.

Overall, the combined plot is more useful for comparison, while the side-by-side plots are better for simplicity and clarity when analyzing each dataset individually.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
week1 = [100, 110, 130, 120, 100, 90, 80]
week2 = [70, 90, 100, 100, 120, 110, 90]

plt.figure(figsize=(10,5))

plt.plot(days, week1, marker='o', label="Week 1")
plt.plot(days, week2, marker='o', label="Week 2")

plt.title("Museum Visitors: Week 1 vs Week 2")
plt.xlabel("Day of the Week")
plt.ylabel("Number of Visitors")

plt.legend()

# Clean look (Storytelling style)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.box(False)

plt.show()
```
